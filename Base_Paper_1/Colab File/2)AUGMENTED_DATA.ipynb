```python
# divide data images into train, test subdirectory
import glob
import random
import shutil

image_dirs = ['ABER', 'AF', 'APB', 'VEB', 'FVN', 'LBBB', 'UCB', 'RBBB', 'S', 'PVC', 'NB', 'PB', 'VF', 'AFIB', 'VB', 'VAL', 'HF', 'HC', 'PAL', 'CMO', 'STA', 'MH', 'BBB', 'DYS', 'MI', 'MYOC', 'USA']

if os.path.isdir('IMG_DATABASE/train/ABER') is False:
    for i in image_dirs:
        current_path = 'IMG_DATABASE/'+i
        path_train = 'IMG_DATABASE/train/'+i
        path_test = 'IMG_DATABASE/test/'+i
        os.makedirs(path_train)
        os.makedirs(path_test)
        path, dirs, files = next(os.walk(current_path))
        no_of_files = len(files)
        no_of_test_dir_files = round(no_of_files*0.2)
        no_of_train_dir_files = no_of_files - no_of_test_dir_files
        print(no_of_files)
        for j in random.sample(glob.glob(current_path+'/fig*'),no_of_train_dir_files):
            shutil.move(j,path_train)
        for j in random.sample(glob.glob(current_path+'/fig*'),no_of_test_dir_files):
            shutil.move(j,path_test)
        
        
# moving 80%,20% data from MIT-BIH_DATABASE/ directory to MIT-BIH_DATABASE/train, MIT-BIH_DATABASE/test subdirectory 

# no need to run this block anymore after data folder (train, test) creation
```

```python
import cv2


def zoom(image, filename):
    count = 0 # keep track of number of images generated
    while count < 5: # repeat until 100 images have been generated
        zoomed = cv2.resize(image, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)
        cropped = zoomed[25:175, 25:175]
        cv2.imwrite(filename[:-4] + 'zoomed' + str(count) + '.png', cropped) # add count to filename
        count += 1 # increment count
    return


def zoom_out(image, filename):
    count = 0
    while count < 5:
        zoomed = cv2.resize(image, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR)
        cv2.imwrite(filename[:-4] + 'zoomed_out' + str(count) + '.png', zoomed)
        count += 1
    return



import numpy as np


def shift(image, filename):
    count = 0
    while count < 5:
        # Generate random shift values between -50 and 50
        x_shift = np.random.randint(-50, 50)
        y_shift = np.random.randint(-50, 50)


        # Shift the image by the random values
        M = np.float32([[1, 0, x_shift], [0, 1, y_shift]])
        shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))


        # Save the shifted image
        cv2.imwrite(filename[:-4] + 'shifted' + str(count) + '.png', shifted)


        count += 1
    return





def shear(image, filename):
    count = 0
    while count < 5:
        # Generate random shear value between -0.5 and 0.5
        shear_val = np.random.uniform(-0.5, 0.5)


        # Apply the shear transformation
        M = np.float32([[1, shear_val, 0], [0, 1, 0]])
        sheared = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))


        # Save the sheared image
        cv2.imwrite(filename[:-4] + 'sheared' + str(count) + '.png', sheared)


        count += 1
    return




def center(image, filename):
    for i in range(4):
        center = image[16:112, 16:112]
        center = cv2.resize(center, (150, 150))
        cv2.imwrite(filename[:-4] + f'centerCenter_{i}.png', center)











```

```python
# data augmentation
augment_dirs = ['ABER/']

for image_dir in augment_dirs:
    path, dirs, files = next(os.walk(os.path.join(train_path,image_dir)))

    
    for file in tqdm(files):
        imagefilepath = os.path.join(train_path,image_dir,file)
        image = cv2.imread(imagefilepath)
        center(image, imagefilepath)
        zoom(image, imagefilepath)
        zoom_out(image, imagefilepath)
        shift(image, imagefilepath)
        shear(image, imagefilepath)
        

# no need to run this block anymore after data augmentation is done
```

```python
import cv2


def zoom(image, filename):
    count = 0 # keep track of number of images generated
    while count < 7: # repeat until 100 images have been generated
        zoomed = cv2.resize(image, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)
        cropped = zoomed[25:175, 25:175]
        cv2.imwrite(filename[:-4] + 'zoomed' + str(count) + '.png', cropped) # add count to filename
        count += 1 # increment count
    return


def zoom_out(image, filename):
    count = 0
    while count < 7:
        zoomed = cv2.resize(image, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR)
        cv2.imwrite(filename[:-4] + 'zoomed_out' + str(count) + '.png', zoomed)
        count += 1
    return



import numpy as np


def shift(image, filename):
    count = 0
    while count < 7:
        # Generate random shift values between -50 and 50
        x_shift = np.random.randint(-50, 50)
        y_shift = np.random.randint(-50, 50)


        # Shift the image by the random values
        M = np.float32([[1, 0, x_shift], [0, 1, y_shift]])
        shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))


        # Save the shifted image
        cv2.imwrite(filename[:-4] + 'shifted' + str(count) + '.png', shifted)


        count += 1
    return





def shear(image, filename):
    count = 0
    while count < 6:
        # Generate random shear value between -0.5 and 0.5
        shear_val = np.random.uniform(-0.5, 0.5)


        # Apply the shear transformation
        M = np.float32([[1, shear_val, 0], [0, 1, 0]])
        sheared = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))


        # Save the sheared image
        cv2.imwrite(filename[:-4] + 'sheared' + str(count) + '.png', sheared)


        count += 1
    return




def center(image, filename):
    for i in range(6):
        center = image[16:112, 16:112]
        center = cv2.resize(center, (150, 150))
        cv2.imwrite(filename[:-4] + f'centerCenter_{i}.png', center)











```
